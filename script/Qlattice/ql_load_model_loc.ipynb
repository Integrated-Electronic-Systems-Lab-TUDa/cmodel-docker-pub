Loading model:

```python

current_model = feyn.Model.load('./models/DataTwo11.json')
current_model.sympify(signif= 3,include_weights=True)
```

```python
current_model.show(update_display=True)
```

Printing some key figures of merrits:

```python
all_data = pd.read_table("../data/id.tbl")
all_data.columns =["v1","v2","v3","v4","id"]
all_data["id"] = all_data["id"].apply(lambda x: x*-1E03)
train, test = train_test_split(all_data,train_size=0.2,random_state=9)

print('qlattice clac of R2:'+ str(current_model.r2_score(test)))
y_pred = current_model.predict(test[['v1','v2','v3','v4']])
y_pred = y_pred.reshape(-1,1)
y_true = test['id'].to_numpy().reshape(-1,1)



utility_func.print_all_metrics(y_true=y_true,y_pred=y_pred,full_name='SR_DataTwo11',error_trsh= 0.05)
```

Plotting the prediction/Response

```python
current_model.plot_response_1d(
    data=test,
    by = 'v2',
    input_constraints = {'v1': 1.10,
                           'v3': -0.40,
                           'v4': -1.60,
                           },
    filename="id-plot"
)
train_plot,_ = train_test_split(train,train_size = 0.01)
test_plot,_ = train_test_split(test,train_size = 0.01)
current_model.plot(data=train_plot,compare_data=test_plot)
```
